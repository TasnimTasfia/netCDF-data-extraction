```python
sylhet_lat = 24.54
sylhet_lon = 91.53
```

```python
rainfall = []
for n in np.arange(0,3654):
  rainfall.append(precip[n,8,0].values.item())
```

```python
rainfall
```

```python
time_numpy_array = time.to_numpy()
time_numpy_array
```

```python
df = pd.DataFrame(time_numpy_array,columns = ['Date'] )
df
```

```python
df['Precipitation (m)'] = rainfall
```

```python
df = df.round(2)
```

```python
df['Precipitation (mm)'] = df['Precipitation (m)']*1000
```
